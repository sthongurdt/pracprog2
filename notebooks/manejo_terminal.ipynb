# Manejo basico de Terminal
Para abrir una terminal en linux podemos dar clic derecho en en el escritorio y dar clic en abrir terminal o presionar las siguientes teclas <code>"ctrl"+"shift"+"t"</code> o abrirla desde las aplicaciones.

Los comandos pueden ser:
- Un programa ejecutable, casi siempre se encuentran en <code>/usr/bin</code>.
- Un comando de utilidad shell.
- Una fucion shell.
- Un alias.

## Comando ls
<code>ls</code> lista el cotenido de la direccion en la que nos encontramos. Para ver todas las opciones del comando escriba en la terminal <code>ls --help</code>

### EJP:
<code>ls -a</code><br>
<code>ls -al</code><br>
<code>ls -alS</code><br>
<code>ls -alSr</code><br>
<code>ls -alSrh</code><br>

## Comando cd
<code>cd</code> permite la navegacion por los diferentes directorios que se encuentran en el sistema.
- Si se escribe solo <code>cd</code> en la terminal nos lleva a la carpeta home (~) del usuario.
- Si se usa <code>cd ..</code> de vuelve al directorio anterior.

### EJP:
<code>cd</code><br>
<code>cd Documents</code><br>
<code>cd ..</code><br>
<code>cd Pictures</code><br>
<code>cd ..</code><br>

## Comando clear
<code>clear</code> limpia la terminal, se usa cuando tenemos demaciado texto. Tambien se puede usar el atajo <code>"ctrl"+"l"</code>

## Comando pwd
<code>pwd</code> nos muestra la ruta/path del directorio en el cual nos encontramos.

### Rutas
- Absolutas: son las rutas que muestra el comando <code>pwd</code>. **EJP:** <code>/home/user/Documents</code> 
- Relativas: son rutas a subdirectorios del directorio en el cual nos encontramos. **EJP:** si estamos en <code>~</code>, la ruta a Documentos seria <code>./Documents</code> o a Imagenes seria <code>./Pictures</code>

## Comando file
<code>file</code> describe el tipo de archivo.

### EJP:
En la carpeta <code>~</code> escribe:
<code>file Documents</code><br>
<code>file Pictures</code><br>

## Comando tree
<code>tree</code> lista el contenido del directorio de una manera mas grafica. <code>tree -L</code> nos lista la cantidad de niveles que queramos ver.

### EJP:
<code>tree</code><br>
<code>tree -L 2</code><br>

## Comando mkdir
<code>mkdir</code> crea directorios. Es **recomendable no crear directorios con espacios para facilitar su manejo.** Es posible crear varios directorios al tiempo separandolo con espacios.

### EJP:
En la carpeta <code>~</code> escribe:
<code>mkdir prb</code><br>
<code>cd prb</code><br>
<code>mkdir prueba</code><br>
<code>mkdir prueba1 prueba2 prueba3</code><br>
<code>ls -al</code><br>

## Comando touch
<code>touch</code> crea un archivo. Es **recomendable no crear directorios con espacios para facilitar su manejo.** Es posible crear varios directorios al tiempo separandolo con espacios.

### EJP:
En la carpeta <code>~</code> escribe:
<code>cd prb</code><br>
<code>touch archivo</code><br>
<code>touch archivo1 archivo2 archivo3</code><br>
<code>ls -al</code><br>

## Comando cp
<code>cp</code> perite la copia de archivos y directorios, para ver todas sus opciones use <code>cp --help</code>. La sintaxis basica del comando es <code>cp /dir/del/archivo/copiado /dir/donde/se/va/a/copiar/el/archivo</code>.

### EJP:
En la carpeta <code>~/prb</code> escribe:
<code>cp archivo file</code>
<code>ls -al</code>
<code>cp archivo prueba/file</code>
<code>ls prueba/file -al</code>

## Comando mv
<code>mv</code> mueve los archivos, la sintaxis es la misma que la de cp. La sintaxis basica del comando es <code>mv /dir/del/archivo/a/moverse /dir/donde/se/va/a/mover/el/archivo</code>.

### EJP:
<code>mv archivo ..</code> mueve el archivo a la carpeta anterior o superior en la jerarquia.
<code>mv archivo1 archivo</code> es una manera de renombrar un archivo

## Comando rm
<code>rm</code> permite eliminar un archivo o directorio, para ver todas sus opciones use <code>rm --help</code>.

### EJP:
<code>rm prb2</code><br>
<code>rm -ri prueba3</code><br>

## Comando head
<code>head</code> muestra las primeras 10 lineas de un archivo. Para ver todas las opciones del comando <code>head</code> use el comando <code>head --help</code>.

### EJP:
En la carpeta <code>~/prb</code> cree un archivo y llenelo a su gusto:
<code>head prb.txt</code>

## Comando tail
<code>tail</code> muestra las ultimas 10 lineas de un archivo. Para ver todas las opciones del comando <code>tail</code> use el comando <code>tail --help</code>.

### EJP:
<code>tail prb.txt</code>

## Comando less
<code>less</code> muestra el contenido del archivo, es mas interactivo y potente que head y tail. Para ver todas las opciones del comando <code>less</code> use el comando <code>less --help</code>.

### EJP:
<code>less prb.txt</code>

## Comando xdg-open
<code>xdg-open</code> permite abrir diferentes archivos en su entorno grafico predeterminado, es posible abrir programas con este comando.

### EJP:
En la carpeta <code>~/prb</code> guarde una imagen:
<code>xdg-open prb.txt</code><br>
<code>xdg-open prb.jpg</code><br>
<code>xdg-open https://www.google.com</code><br>
**NOTA:** si el proceso sigue activo en la terminal o que no permite escribir nuevas lineas, usa la combinacion de teclas <code>ctrl + c</code>. Tambien se puese agregar al final del comando <code>& disown</code>.
<code>xdg-open prb.jpg & disown</code><br>

## Comando type
<code>type</code> muestra el tipo de comando que se usa o la direccion donde podemos encontrarlo.

### EJP:
<code>type cd</code><br>
<code>type mkdir</code><br>
<code>type ls</code><br>

## Creacion de un alias
Los alias se pueden crear segun nos convengan.

### EJP:
<code>alias lst="ls -alh"</code><br>
<code>lst</code><br>
**Nota:** si el alias se crea de esta manera solo se podra utilizar mientras la terminal este abierta.

## Wildcards
Esto nos permite la busqueda de archivos especificos, pueden ser <code>*</code> o <code>?</code>. El <code>*</code> toma los archivos que tengan cualquier coincidencia, el <code>?</code> cuenta como un solo caracter y si se desea buscar con mas caracteres se debe colocar la misma cantidad de <code>?</code>. 

### EJP:
En la carpeta <code>~/prb</code>:
<code>ls *.jp*</code><br>
<code>ls prb*</code><br>
<code>ls prb?</code> no nos muestra nada<br>
<code>ls prb????</code><br>
<code>ls [[:upper:]]*</code> lista todo el contenido que inicie con mayusculas<br>
<code>ls [[:lower:]]*</code> lista todo el contenido que inicie con minusculas<br>
<code>ls -d [[:lower:]]*</code> lista todas las carpetas/directorios que inicie con minusculas<br>
<code>ls [n]*</code> lista todo el contenido que inicie con n<br>
<code>ls [p]*</code> lista todo el contenido que inicie con p<br>
<code>ls [n]*</code> lista todo el contenido que inicie con n<br>
<code>ls *[[:digit:]]</code> lista todo el contenido que termine en un numero<br>

## Redirecciones
Cuando se usan comandos siempre vamos a dar al menos un valor de entrada (stdin(0)) y en consecuencia podemos tener una salida (stdout(1)) o un error (stderr(2)). La sintaxis es la siguiente: <code>/ruta/comando-ejecutable > /ruta/del/archivo/de/redireccion</code>. 

Los resultados de esta sintaxis crea un archivo cada ves que se ejecuta. Si lo que se desea es concatenar multiples redirecciones use la sintaxis: <code>/ruta/comando-ejecutable >> /ruta/del/archivo/de/redireccion</code>

### EJP:
En la carpeta <code>~/prb</code>:
<code>ls > lsprb.txt</code>
<code>less lsprb.txt</code>
<code>ls -alh > lsprb.txt</code>
<code>cat lsprb.txt</code>
<code>ls asdfg >> error.txt</code>
<code>less error.txt</code> No muestra nada.

Si desea que cualquier salida sea almacenada en la redireccion agregue <code>2>&1</code>
<code>ls qwert > error.txt 2>&1</code>
<code>cat error.txt 2>&1</code>
